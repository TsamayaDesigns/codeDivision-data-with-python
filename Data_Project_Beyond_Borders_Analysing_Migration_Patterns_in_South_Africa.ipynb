## The Dataset

I will be using the Talent Migration Dataset. The metrics and data are a product of the partnership between the World Bank and LinkedIn, and cover industry employment, skill, and migration metrics for more than 100 countries.

**Location:**
* URL: "https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true"

**Contents:**
* Worksheets:
  1. Country Migration
  2. Industry Migration
  3. Skill Migration

**Code:**

In the code section below, a function **"get_excel_data()"** is created that will:
  1. Specify the location (URL) of the dataset
  2. Specify the new dataframes for each separate worksheet
    * country_df
    * industry_df
    * skill_df
  3. Read the specified worksheets (Worksheets) into its own dataframe (2)
  4. Display the column names and the first 5 rows of each dataframe (2)
  5. Return a separate dataframe for each worksheet

**Display**

Displayed directly below the Code Section (per dataframe):
* Dataframe name
* Column names
* First 5 rows of data

---



```python
import pandas as pd
pd.set_option('display.width', 270)

# Import CSV-data from GitHub
def get_excel_data():
  url = "https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true"
  country_df = pd.read_excel(url, sheet_name = "Country Migration")
  industry_df = pd.read_excel(url, sheet_name = "Industry Migration")
  skill_df =  pd.read_excel(url, sheet_name = "Skill Migration")

  # Display Datasets (first 5 rows)
  print("Country Dataframe: ")
  display(country_df.head())
  print("\nIndustry Dataframe: ")
  display(industry_df.head())
  print("\nSkill Dataframe: ")
  display(skill_df.head())

  return country_df, industry_df, skill_df

country_df, industry_df, skill_df = get_excel_data()
```

```python
# Describing the data
def describe_data():
  # 1. Country Migration
  print("\n1. Country Migration Dataframe")
  print("------------------------------")
  print("\nNumber of Rows & Columns: \n", country_df.shape)
  print("\nColumn Names, Non-null Counts & Datatypes: ")
  print(country_df.info())
  print("\nNumber of (unique) Countries: \n", len(country_df["base_country_code"].unique()))
  print("--------------------------------------------")

  # 2. Industry Migration
  print("\n2. Industry Migration Dataframe")
  print("-------------------------------")
  print("\nNumber of Rows & Columns: \n", industry_df.shape)
  print("\nColumn Names, Non-null Counts & Datatypes: ")
  print(industry_df.info())
  print("\nNumber of (unique) Countries: \n", len(industry_df["country_code"].unique()))
  print("--------------------------------------------")

  # 3. Skill Migration
  print("\n3. Skill Migration Dataframe")
  print("----------------------------")
  print("\nNumber of Rows & Columns: \n", skill_df.shape)
  print("\nColumn Names, Non-null Counts & Datatypes: ")
  print(skill_df.info())
  print("\nNumber of (unique) Countries: \n", len(skill_df["country_code"].unique()))
  print("--------------------------------------------")

  # 4. Check for duplicates in the different dataframes
  print("\n4. Duplicates in dataframes?")
  print("----------------------------")
  print("\nCountry dataframe:", country_df.duplicated().sum())
  print("Industry dataframe:", industry_df.duplicated().sum())
  print("Skill dataframe:", skill_df.duplicated().sum())
  print("--------------------------------------------")

  # 5. Find the Countries in country_df but not in industry_df
  print("\n4. Differences in dataframes: 'Country Code'?")
  print("---------------------------------------------")
  country_codes_country_df = set(country_df["base_country_code"].unique())
  country_codes_industry_df = set(industry_df["country_code"].unique())
  country_codes_skill_df = set(skill_df["country_code"].unique())
  countries_in_country_not_in_skill = country_codes_country_df - country_codes_skill_df
  countries_in_country_not_in_industry = country_codes_country_df - country_codes_industry_df
  countries_in_skill_not_in_industry = country_codes_skill_df - country_codes_industry_df

 # Print the results
  print("Countries in Country dataframe & not in Skill dataframe: ")
  print(len(countries_in_country_not_in_skill))
  print("Countries in Country dataframe & not in Industry dataframe: ")
  print(countries_in_country_not_in_industry)
  print("Countries in Skill dataframe & not in Industry dataframe: ")
  print(countries_in_skill_not_in_industry)
describe_data()
## describe the dataset (number of records, statistics, columns, etc)
```
